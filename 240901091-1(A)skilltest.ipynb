```python
scores = [78, 45, 62, 89, 55, 34, 91, 73, 68, 49,84, 52, 47, 66, 71, 59, 38, 95, 81, 60]
average_score = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)
print("\nOverall Class Performance")
print("Average Score:", average_score)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)
print("\nPass/Fail Classification (Pass mark = 50)")
for i, score in enumerate(scores, start=1):
    if score >= 50:
        print(f"Student {i}: {score} - Pass")
    else:
        print(f"Student {i}: {score} - Fail")
def assign_grade(score):
    if score >= 85:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
print("\nGrades")
for i, score in enumerate(scores, start=1):
    grade = assign_grade(score)
    print(f"Student {i}: Score = {score}, Grade = {grade}")
```
