```python
import numpy as np
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
temp=arr[[0,1,2,3],[3,2,1,0]]
print("\nElements at indices(0,3),(1,2),(2,1),""(3,0):\n",temp)
```

```python
cond=arr>2
temp=arr[cond]
print("\nElements grater than 2:\n",temp)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print("\nOrignal arrays:\n",arr1,arr2)print("\nJoined array:\n",arr)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print("\nOrignal arrays:\n",arr1,arr2)
print("\nJoined array:\n",arr)
```

```python
arr=np.hstack((arr1,arr2))
print("\nHorizontal joining:\n",arr)
```

```python
arr=np.vstack((arr1,arr2))
print("\nVertical joining:\n",arr)
```

```python
arr=np.dstack((arr1,arr2))
print("\nDepth joining:\n",arr)
```

```python
arr=np.array([1,2,3,4,5,6])
newarr=np.array_split(arr,3)
print("\nOrignal array:\n",arr)
print("\nSplitted array:\n",newarr)
print("\nSplitted array in another form:\n")
print(newarr[0])
print(newarr[1])
print(newarr[2])

```
